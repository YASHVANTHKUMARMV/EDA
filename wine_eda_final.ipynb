```python
import pandas as pd
df = pd.read_csv("winequality-red.csv")
df.head()
```

```python
# information about the dataset
df.info()
```

```python
print("Shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
df.info()
display(df.describe())
```

```python
df.duplicated()
df.drop_duplicates(inplace=True)
```

```python
# 4. Univariate Analysis - Histograms
for col in df.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30, color='crimson')
    plt.title(f"Distribution of {col}")
    plt.tight_layout()
    plt.show()
```

```python
# 5. Boxplots for Outlier Detection
for col in df.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col], color='orange')
    plt.title(f"Boxplot of {col}")
    plt.tight_layout()
    plt.show()
```

```python

# 6. Correlation Matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# 7. Pairwise Relationships
selected_cols = ['alcohol', 'volatile acidity', 'sulphates', 'citric acid', 'quality']
sns.pairplot(df[selected_cols], hue='quality', palette='Set2')
plt.show()
```

```python
# 8. Boxplots of Features by Wine Quality
for col in df.columns[:-1]:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='quality', y=col, data=df, palette='Set3')
    plt.title(f"{col} vs Quality")
    plt.tight_layout()
    plt.show()
```

```python
# 9. Quality Value Counts (Target Balance)
sns.countplot(x='quality', data=df, palette='pastel')
plt.title("Distribution of Wine Quality")
plt.show()
```

```python
# 10. Convert Quality to Binary for Classification (Optional)
df['good_quality'] = [1 if q >= 7 else 0 for q in df['quality']]

sns.countplot(x='good_quality', data=df, palette='Set1')
plt.title("Good vs Bad Quality Wine")
plt.xticks([0, 1], ['Bad', 'Good'])
plt.show()
```
